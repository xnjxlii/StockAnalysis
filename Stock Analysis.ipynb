```python
#Import needed libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Download Google Data
yf.download('GOOG',
            start='2019-03-31',end='2024-03-31',interval='1d')
```

```python
#Download Microsoft Data
yf.download('MSFT',
            start='2019-03-31',end='2024-03-31',interval='1d')
```

```python
#Download Microsoft Data
yf.download('AMZN',
            start='2019-03-31',end='2024-03-31',interval='1d')
```

```python
#Download Amazon Data and store in amzn_data
amzn_data = yf.download('AMZN', start='2019-03-31', end='2024-03-31', interval='1d')

# Calculate the monthly average closing price for Amazon
monthly_avg_close_amzn = amzn_data['Close'].resample('M').mean()

# Calculate the year-over-year (YoY) percentage change in the closing price for Amazon
yearly_close_amzn = amzn_data['Close'].resample('Y').last()
yearly_close_change_amzn = yearly_close_amzn.pct_change() * 100

# Identify the stock with the highest average volume over the period
avg_volume_amzn = amzn_data['Volume'].mean()

# Print results
print("Monthly Average Closing Price for Amazon (AMZN):")
print(monthly_avg_close_amzn)
print("\nYear-over-Year (YoY) Percentage Change in Closing Price for Amazon (AMZN):")
print(yearly_close_change_amzn)
print("\nAverage Volume for Amazon (AMZN):", avg_volume_amzn)

```

```python
#Data stored in msft_data
msft_data = yf.download('MSFT', start='2019-03-31', end='2024-03-31', interval='1d')

# Calculate the monthly average closing price for Microsoft
monthly_avg_close_msft = msft_data['Close'].resample('M').mean()

# Calculate the year-over-year (YoY) percentage change in the closing price for Microsoft
yearly_close_msft = msft_data['Close'].resample('Y').last()
yearly_close_change_msft = yearly_close_msft.pct_change() * 100

# Identify the stock with the highest average volume over the period
avg_volume_msft = msft_data['Volume'].mean()

# Print results
print("Monthly Average Closing Price for Microsoft(MSFT):")
print(monthly_avg_close_msft)
print("\nYear-over-Year (YoY) Percentage Change in Closing Price for Microsoft(MSFT):")
print(yearly_close_change_msft)
print("\nAverage Volume for Microsoft (MSFT):", avg_volume_msft)

```

```python
#Data stored in google_data
google_data= yf.download('GOOG', start='2019-03-31', end='2024-03-31', interval='1d')

# Calculate the monthly average closing price for google
monthly_avg_close_google = google_data['Close'].resample('M').mean()

# Calculate the year-over-year (YoY) percentage change in the closing price for GOOGLE
yearly_close_google = google_data['Close'].resample('Y').last()
yearly_close_change_google = yearly_close_google.pct_change() * 100

# average volume over the period
avg_volume_google = google_data['Volume'].mean()

# Print results
print("Monthly Average Closing Price for Google(GOOG):")
print(monthly_avg_close_google)
print("\nYear-over-Year (YoY) Percentage Change in Closing Price for Google(GOOG):")
print(yearly_close_change_google)
print("\nAverage Volume for Google(GOOG):", avg_volume_google)
```

```python
avg_volume_msft = msft_data['Volume'].mean()
avg_volume_google = google_data['Volume'].mean()
avg_volume_amzn = amzn_data['Volume'].mean()

# Find the stock with the highest average volume
highest_avg_volume_stock = max([(avg_volume_msft, 'MSFT'), (avg_volume_google, 'GOOG'), (avg_volume_amzn, 'AMZN')])
print("Stock with the highest average volume:", highest_avg_volume_stock[1], "(", highest_avg_volume_stock[0], ")")

#Additional analysis:Calculating the expected rate of return

#Calculate daily returns
tickers = ['AMZN', 'GOOG', 'MSFT']
stock_data = yf.download(tickers, start='2019-03-31', end='2024-03-31', interval='1d')
daily_returns = stock_data['Close'].pct_change()

#Compute the average daily return
expected_daily_return = daily_returns.mean()

#Annualize the expected daily return
# Assuming 252 trading days in a year
trading_days_per_year = 252 
expected_annual_return = expected_daily_return * trading_days_per_year

#Display expected annual return 
print("\nExpected Annual Rate of Return:")
print(expected_annual_return)
```

Results: Microsoft has a higher expected annual rate of retrun so investors can invest in microsoft rather than Amazon and Google.

```python
# Line plot showing the closing prices of the stocks over time
closing_prices = stock_data['Close']
plt.figure(figsize=(12, 6))
closing_prices.plot()
plt.title('Closing Prices of Stocks Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend(tickers)
plt.grid(True)
plt.show()
```

```python
# Bar chart comparing the average closing prices of the stocks
avg_closing_prices = closing_prices.mean()
plt.figure(figsize=(10, 6))
avg_closing_prices.plot(kind='bar', color='skyblue')
plt.title('Average Closing Prices of Stocks')
plt.xlabel('Stocks')
plt.ylabel('Average Closing Price (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Heatmap showing the correlation between the daily price movements of the stocks
plt.figure(figsize=(8, 6))
sns.heatmap(closing_prices.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Daily Price Movements')
plt.show()
```

```python
#Individual histograms to analyse the daily returns of each stock
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, ticker in enumerate(tickers, 1):
    plt.subplot(2, 2, i)
    daily_returns = stock_data['Close'][ticker].pct_change().dropna()
    sns.histplot(daily_returns, bins=50, kde=True, label=ticker, alpha=0.5)
    plt.title(f'Histogram of Daily Returns for {ticker}')
    plt.xlabel('Daily Return')
    plt.ylabel('Frequency')
    plt.legend()

plt.tight_layout()
plt.show()
```

By comparing the histograms of different stocks,investors can assess their relative risk and return characteristics. Stocks with wider or more dispersed histograms may be perceived as riskier, while those with narrower and more concentrated histograms may be considered safer investments.
- Here it is noticed that MSFT is less riskier than AMZN and GOOG as it has a narrow histogram.

```python
plt.figure(figsize=(10, 6))

# Define a list of colors for each stock,easier to differentiate
colors = ['blue', 'green', 'red', 'orange']

for i, ticker in enumerate(tickers):
    daily_returns = stock_data['Close'][ticker].pct_change().dropna()
    sns.histplot(daily_returns, bins=50, kde=True, label=ticker, color=colors[i], alpha=0.5)

plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

A taller distribution as in MSFT with higher peak suggests that returns are more concentrated around certain values. This indicates lower volatility.
